```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import statsmodels.api as sm
import yfinance as yf
import datetime as dt
```

# **Import Data**

```python
def getData(stock, start, end):
    stockData = yf.download(stock, start, end, auto_adjust=True)
    stockData = stockData['Close']
    returns = stockData.pct_change()
    return returns

Nifty = ['^NSEI']
ExposedStockList = ['ARVIND.NS', 'VTL.NS', 'WELSPUNLIV.NS', 'TITAN.NS', 'RAJESHEXPO.NS', 'KALYANKJIL.NS', 'BATAINDIA.NS', 'MIRZAINT.NS', 'RELAXO.NS', 'AVANTIFEED.NS', 'WATERBASE.BO', 'APEX.NS', 'AARTIIND.NS', 'SRF.NS', 'DEEPAKNTR.NS', 'BHARATFORG.NS', 'MOTHERSON.NS', 'SUNDRMFAST.NS']
NonExposedStockList = ['SUNPHARMA.NS', 'DRREDDY.NS', 'CIPLA.NS', 'TATAELXSI.NS', 'DIXON.NS', 'ASMTEC.BO', 'RELIANCE.NS', 'ONGC.NS', 'COALINDIA.NS']
# stockList = [stock + '.NS' for stock in stockList]
endDate = dt.datetime.now()
startDate = endDate - dt.timedelta(days=900)
```

```python
dfN = getData(Nifty, startDate, endDate)
df1 = getData(ExposedStockList, startDate, endDate)
df2 = getData(NonExposedStockList, startDate, endDate)
```

# **Calculating Abnormal Returns for Event 1 (April 2nd)**

# Tariff Exposed Firms

```python
event_date = pd.Timestamp("2025-04-02")

# Re-establish estimation_window and event_window for df1 (Exposed Firms)
# This is necessary because these variables were overwritten by the execution of the non-exposed firms section.
exposed_firms_event_loc = df1.index.get_loc(event_date)
nifty_event_loc = dfN.index.get_loc(event_date)

exposed_firms_estimation_window = df1.iloc[exposed_firms_event_loc-250:exposed_firms_event_loc-30]
exposed_firms_event_window = df1.iloc[exposed_firms_event_loc-1:exposed_firms_event_loc+2]

nifty_estimation_window = dfN.iloc[nifty_event_loc-250:nifty_event_loc-30]
nifty_event_window = dfN.iloc[nifty_event_loc-1:nifty_event_loc+2]

Exposed_AR_matrix = pd.DataFrame(index=exposed_firms_event_window.index) # Corrected index source

X = nifty_estimation_window["^NSEI"]
X = sm.add_constant(X)

for firm in ExposedStockList:
  Y = exposed_firms_estimation_window[firm]
  X = nifty_estimation_window["^NSEI"]
  X = sm.add_constant(X)
  model = sm.OLS(Y, X).fit()
  alpha = model.params["const"]
  beta = model.params["^NSEI"]

  X_event = sm.add_constant(nifty_event_window["^NSEI"])
  expected = model.predict(X_event)
  # print(f"{firm}: {expected}")
  Exposed_AR_matrix[firm] = exposed_firms_event_window[firm] - expected
  actual = exposed_firms_event_window[firm]
  abnormal = actual - expected
  CAR = abnormal.sum()
  # print(f"{firm}:{CAR}")


Exposed_AAR = Exposed_AR_matrix.mean(axis=1)
Exposed_AAR
ExposedCAAR = Exposed_AAR.sum()
ExposedCAAR
```

# Non Tariff Exposed Firms

```python
event_date = pd.Timestamp("2025-04-02")

# Re-establish estimation_window and event_window for df1 (Exposed Firms)
# This is necessary because these variables were overwritten by the execution of the non-exposed firms section.
non_exposed_firms_event_loc = df2.index.get_loc(event_date)
nifty_event_loc = dfN.index.get_loc(event_date)

non_exposed_firms_estimation_window = df2.iloc[non_exposed_firms_event_loc-250:non_exposed_firms_event_loc-30]
non_exposed_firms_event_window = df2.iloc[non_exposed_firms_event_loc-1:non_exposed_firms_event_loc+2]

nifty_estimation_window = dfN.iloc[nifty_event_loc-250:nifty_event_loc-30]
nifty_event_window = dfN.iloc[nifty_event_loc-1:nifty_event_loc+2]

Non_Exposed_AR_matrix = pd.DataFrame(index=non_exposed_firms_event_window.index)

X = nifty_estimation_window["^NSEI"]
X = sm.add_constant(X)

for firm in NonExposedStockList:
  Y = non_exposed_firms_estimation_window[firm]

  model = sm.OLS(Y, X).fit()
  alpha = model.params["const"]
  beta = model.params["^NSEI"]

  X_event = sm.add_constant(nifty_event_window["^NSEI"])
  expected = model.predict(X_event)
  # print(f"{firm}: {expected}")
  Non_Exposed_AR_matrix[firm] = non_exposed_firms_event_window[firm] - expected

  actual = non_exposed_firms_event_window[firm] # Corrected variable
  abnormal = actual - expected
  CAR = abnormal.sum()
  # print(f"{firm}:{CAR}")

Non_Exposed_AAR = Non_Exposed_AR_matrix.mean(axis=1)
NonExposedCAAR = Non_Exposed_AAR.sum()
NonExposedCAAR
```

# Graphing Exposed vs Non Exposed for April 2nd

# **Calculating Abnormal Returns for Event 4 (February 2nd)**

# Tariff Exposed Firms

```python
event_date = pd.Timestamp("2026-02-02")
exposed_firms_event_loc = df1.index.get_loc(event_date)
nifty_event_loc = dfN.index.get_loc(event_date)

exposed_firms_estimation_window = df1.iloc[exposed_firms_event_loc-250:exposed_firms_event_loc-30]
exposed_firms_event_window = df1.iloc[exposed_firms_event_loc-1:exposed_firms_event_loc+2]

nifty_estimation_window = dfN.iloc[nifty_event_loc-250:nifty_event_loc-30]
nifty_event_window = dfN.iloc[nifty_event_loc-1:nifty_event_loc+2]

Exposed_AR_matrix = pd.DataFrame(index=exposed_firms_event_window.index)

for firm in ExposedStockList:
  Y_firm_est = exposed_firms_estimation_window[firm]
  X_nifty_est = nifty_estimation_window["^NSEI"]

  # Combine firm returns and Nifty returns for the estimation window to ensure common index
  temp_df_est = pd.DataFrame({'Y': Y_firm_est, 'X_nifty': X_nifty_est}).dropna()

  if temp_df_est.empty or len(temp_df_est) < 2:
    # print(f"Warning: Insufficient aligned data for firm {firm} in estimation window. Skipping.")
    continue

  Y_ols = temp_df_est['Y']
  X_ols = sm.add_constant(temp_df_est['X_nifty'])

  model = sm.OLS(Y_ols, X_ols).fit()
  # alpha = model.params["const"]
  # beta = model.params["^NSEI"]

  # Prepare data for prediction in the event window, ensuring index alignment
  Y_firm_event = exposed_firms_event_window[firm]
  X_nifty_event = nifty_event_window["^NSEI"]

  temp_df_event = pd.DataFrame({'Y_event': Y_firm_event, 'X_nifty_event': X_nifty_event}).dropna()

  if temp_df_event.empty:
    print(f"Warning: No aligned data for firm {firm} in event window. Skipping AR calculation.")
    continue

  X_event_pred = sm.add_constant(temp_df_event['X_nifty_event'])

  expected = model.predict(X_event_pred)

  actual = temp_df_event['Y_event'].reindex(expected.index) # Reindex actual to match expected's index

  # Assign abnormal returns to the AR_matrix
  # Ensure the index of the result aligns with Exposed_AR_matrix's index
  Exposed_AR_matrix.loc[actual.index, firm] = actual - expected

  # abnormal = actual - expected
  # CAR = abnormal.sum()
  # print(f"{firm}:{CAR}")


Exposed_AAR = Exposed_AR_matrix.mean(axis=1)
Exposed_AAR
ExposedCAAR = Exposed_AAR.sum()
ExposedCAAR
```
